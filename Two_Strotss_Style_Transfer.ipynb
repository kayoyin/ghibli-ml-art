```python
#@title Helper Functions
import torch
import requests
import PIL.Image
from io import BytesIO
import matplotlib.pyplot as plt

device = 'cuda' if torch.cuda.is_available() else 'cpu'
if not torch.cuda.is_available():
    print('YOU ARE NOT USING A GPU.  IT\'S GONNA BE REAAALLLLY SLOW')
    print('Go to the top of the page.  Click Runtime -> Change Runtime Type -> Hardware accelerator')
    print('From the dropdown, select GPU and rerun all this stuff')

def pil_loader_internet(url):
    response = requests.get(url)
    img = PIL.Image.open(BytesIO(response.content))
    return img.convert('RGB')

# Added
def pil_loader_local(path):
    img = PIL.Image.open(path)
    return img.convert('RGB')

def show_img(img):
    # Code for displaying at actual resolution from:
    # https://stackoverflow.com/questions/28816046/displaying-different-images-with-actual-size-in-matplotlib-subplot
    dpi = 80
    height, width, depth = img.shape
    figsize = width / float(dpi), height / float(dpi)
    plt.figure(figsize=figsize)

    plt.imshow(img)
    plt.xticks([])
    plt.yticks([])
    plt.show()


def plot_style_and_content(style, content):
    fig, ax = plt.subplots(1,2, figsize=(10,5))
    ax[0].imshow(content)
    ax[0].set_xticks([])
    ax[0].set_yticks([])
    ax[0].set_title('Content')
    ax[1].imshow(style)
    ax[1].set_xticks([])
    ax[1].set_yticks([])
    ax[1].set_title('Style')
    plt.show()
```

```python
# Overwrite fn
def calculate_loss(feat_result, feat_content, feat_style1, feat_style2, indices, content_weight, moment_weight=1.0):
    # spatial feature extract
    num_locations = 1024
    # spatial_result: does not depend on feat_style?
    spatial_result, spatial_content = spatial_feature_extract(feat_result, feat_content, indices[0][:num_locations], indices[1][:num_locations])
    loss_content = content_loss(spatial_result, spatial_content)

    feat_max = 3+2*64+128*2+256*3+512*2 # (sum of all extracted channels)

    # STYLE 1
    d = feat_style1.shape[1]
    spatial_style1 = feat_style1.view(1, d, -1, 1)

    loss_remd1 = style_loss(spatial_result[:, :feat_max, :, :], spatial_style1[:, :feat_max, :, :])
    loss_moment1 = moment_loss(spatial_result[:,:-2,:,:], spatial_style1, moments=[1,2]) # -2 is so that it can fit?

    # palette matching
    content_weight_frac = 1./max(content_weight,1.)
    loss_moment1 += content_weight_frac * style_loss(spatial_result[:,:3,:,:], spatial_style1[:,:3,:,:])

    loss_style1 = loss_remd1 + (moment_weight/2) * loss_moment1
    # print(f'Style: {loss_style.item():.3f}, Content: {loss_content.item():.3f}')

    # STYLE 2
    d = feat_style2.shape[1]
    spatial_style2 = feat_style2.view(1, d, -1, 1)

    loss_remd2 = style_loss(spatial_result[:, :feat_max, :, :], spatial_style2[:, :feat_max, :, :])
    loss_moment2 = moment_loss(spatial_result[:,:-2,:,:], spatial_style2, moments=[1,2]) # -2 is so that it can fit?

    # palette matching
    content_weight_frac = 1./max(content_weight,1.)
    loss_moment2 += content_weight_frac * style_loss(spatial_result[:,:3,:,:], spatial_style2[:,:3,:,:])
    
    loss_style2 = loss_remd2 + (moment_weight/2) * loss_moment2
    # print(f'Style: {loss_style.item():.3f}, Content: {loss_content.item():.3f}')
    
    # COMBINE
    style_weight = 1.0 + moment_weight
    loss_total = (content_weight * loss_content + loss_style1 + loss_style2) / (content_weight + style_weight)
    #loss_total = (content_weight * loss_content + loss_style2) / (content_weight + style_weight)
    return loss_total
```

```python
# Overwrite fn
def optimize(result, content, style1, style2, scale, content_weight, lr, extractor):
    # torch.autograd.set_detect_anomaly(True)
    result_pyramid = make_laplace_pyramid(result, 5)
    result_pyramid = [l.data.requires_grad_() for l in result_pyramid]

    opt_iter = 200
    # if scale == 1:
    #     opt_iter = 800

    # use rmsprop
    optimizer = optim.RMSprop(result_pyramid, lr=lr)

    # extract features for content
    feat_content = extractor(content) # 

    stylized = fold_laplace_pyramid(result_pyramid)
    # let's ignore the regions for now
    # some inner loop that extracts samples
    feat_style1 = None
    for i in range(5):
        with torch.no_grad():
            # r is region of interest (mask)
            feat_e = extractor.forward_samples_hypercolumn(style1, samps=1000)
            feat_style1 = feat_e if feat_style1 is None else torch.cat((feat_style1, feat_e), dim=2)
    feat_style1.requires_grad_(False)

    feat_style2 = None
    for i in range(5):
        with torch.no_grad():
            # r is region of interest (mask)
            feat_e = extractor.forward_samples_hypercolumn(style2, samps=1000)
            feat_style2 = feat_e if feat_style2 is None else torch.cat((feat_style2, feat_e), dim=2)
    feat_style2.requires_grad_(False)

    # init indices to optimize over
    print(feat_style1.shape[1])
    print(sample_indices(feat_content[0], feat_style1))
    # PROBLEM HERE?
    xx, xy = sample_indices(feat_content[0], feat_style1) # 0 to sample over first layer extracted
    for it in range(opt_iter):
        optimizer.zero_grad()

        stylized = fold_laplace_pyramid(result_pyramid)
        # original code has resample here, seems pointless with uniform shuffle
        # ...
        # also shuffle them every y iter
        if it % 1 == 0 and it != 0:
            np.random.shuffle(xx)
            np.random.shuffle(xy)
        feat_result = extractor(stylized)

        loss = calculate_loss(feat_result, feat_content, feat_style1, feat_style2, [xx, xy], content_weight)
        loss.backward()
        optimizer.step()
    return stylized
```

```python
# Overwrite fn
def strotss(content_pil, style1_pil, style2_pil, content_weight=1.0*16.0, device='cuda:0', space='uniform'):
    content_np = pil_to_np(content_pil)
    style1_np = pil_to_np(style1_pil)
    style2_np = pil_to_np(style2_pil)
    
    content_full = np_to_tensor(content_np, space).to(device)
    style1_full = np_to_tensor(style1_np, space).to(device)
    style2_full = np_to_tensor(style2_np, space).to(device)

    lr = 2e-3
    extractor = Vgg16_Extractor(space=space).to(device)

    scale_last = max(content_full.shape[2], content_full.shape[3])
    scales = []
    for scale in range(10):
        divisor = 2**scale
        if min(content_pil.width, content_pil.height) // divisor >= 33:
            scales.insert(0, divisor)
    
    clow = -1.0 if space == 'uniform' else -1.7
    chigh = 1.0 if space == 'uniform' else 1.7

    for scale in scales:
        # rescale content to current scale
        content = tensor_resample(content_full, [ content_full.shape[2] // scale, content_full.shape[3] // scale ])
        style1 = tensor_resample(style1_full, [ style1_full.shape[2] // scale, style1_full.shape[3] // scale ])
        style2 = tensor_resample(style2_full, [ style2_full.shape[2] // scale, style2_full.shape[3] // scale ])
        print(f'Optimizing at resolution [{content.shape[2]}, {content.shape[3]}]')

        # upsample or initialize the result
        if scale == scales[0]:
            # first
            # RANDOM INIT
            #result = torch.rand(content.shape).to(device)
            # MODIFIED ORIGINAL INIT
            result = laplacian(content) + style1.mean(2,keepdim=True).mean(3,keepdim=True) + style2.mean(2,keepdim=True).mean(3,keepdim=True)
        elif scale == scales[-1]:
            # last 
            result = tensor_resample(result, [content.shape[2], content.shape[3]])
            lr = 1e-3
        else:
            result = tensor_resample(result, [content.shape[2], content.shape[3]]) + laplacian(content)

        # do the optimization on this scale
        result = optimize(result, content, style1, style2, scale, content_weight=content_weight, lr=lr, extractor=extractor)

        # Show intermediate result
        result_image = tensor_to_np(torch.clamp(result, clow, chigh)) # 
        # renormalize image
        result_image -= result_image.min()
        result_image /= result_image.max()
        show_img(result_image)

        # next scale lower weight
        content_weight /= 2.0

    result_image = tensor_to_np(tensor_resample(torch.clamp(result, clow, chigh), [content_full.shape[2], content_full.shape[3]])) # 
    # renormalize image
    result_image -= result_image.min()
    result_image /= result_image.max()
    return np_to_pil(result_image * 255.)
```

```python
#@title Run Two-STROTSS
#@markdown Put paths to images for content_path, style1_url, and style2_url
content_path = '../blank_white.jpg'  #@param {type:"string"}
style1_path = '../totoro040.jpg'  #@param {type:"string"}
style2_path = '../ponyo050.jpg'  #@param {type:"string"}

content_pil = pil_loader_local(content_path)
style1_pil = pil_loader_local(style1_path)
style2_pil = pil_loader_local(style2_path)

plot_style_and_content(style1_pil, content_pil)
plot_style_and_content(style2_pil, content_pil)

max_width = 1028  #@param {type:"integer"}

#@markdown How much weight to give to content. 
content_weight =  0#@param {type:"number"}
content_weight *= 16.0 

result = strotss(pil_resize_long_edge_to(content_pil, max_width), 
        pil_resize_long_edge_to(style1_pil, max_width),
        pil_resize_long_edge_to(style2_pil, max_width), 
        content_weight, device, "vgg")
print('Final Result')
show_img(pil_to_np(result))
```
